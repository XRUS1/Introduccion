# Uso de Toon Boom Harmony

 Al usar **Toon Boom Harmony** para las ilustraciones del papel que se levantan, los alumnos le darán al libro **pop-up físico** una **identidad visual cinematográfica y de autor espectacular**. Además, la idea de **incluir en cada página una dirección web que active tanto la Realidad Aumentada (RA) como la Realidad Virtual (RV)** es el núcleo puro del **WebXR**. Permite crear un ecosistema transmedia donde el usuario decide cómo consumir la experiencia.

 ## 1. El Rol de Toon Boom Harmony: De la Pantalla al Papel

 Los alumnos no solo usarán **Toon Boom para animar en la web**, sino también como su **herramienta de ilustración editorial**.<br>

- **El Flujo:** Ellos dibujarán las fachadas de las casas, los carros y los personajes en **Toon Boom** aprovechando sus pinceles vectoriales y paletas de color avanzadas.
  
- **La Impresión:** Exportarán estos elementos en alta resolución (capas planas en formatos .png o .tiff a 300 DPI) para **imprimirlos**, **recortarlos** y **armar las maquetas físicas del pop-up**.

- **El Secreto del Tracking:** Esas mismas ilustraciones de **Toon Boom impresas en el papel** serán los **"marcadores visuales"** que la cámara del celular reconocerá para activar la magia digital.


## 2. Cómo Estructurar la Dirección Web en cada Página

Para que el acceso a la web sea inmediato y no rompa la experiencia de lectura, usar un diseño de interfaz física (UI análoga):

- **Código QR + URL corta:** Cada página del libro debe incluir un código QR discreto pero legible, acompañado de la URL corta de su GitHub Pages (ej. usuario.github.io/proyecto/pagina1).

- **Diseño Integrado:** El código QR no debe parecer un parche. Los alumnos deben **integrarlo orgánicamente en la ilustración de Toon Boom** (por ejemplo, **el QR puede ser un cuadro colgado en la pared de la casa pop-up**, o un **espectacular espectacular en la calle por donde pasa el carro**).

## 3. La Experiencia Dual: RA y RV (WebXR + A-Frame)

Dado que estás utilizando **A-Frame**, esta librería es perfecta para resolver tu idea porque tiene soporte nativo para **Realidad Mixta (XR)** en el navegador. Cuando el usuario escanee el código QR con su celular, la web debe ofrecer dos modos claros:

- **Modo A: Realidad Aumentada** (Mirando el libro), Al abrir la web en el celular, se activa la cámara trasera. Al apuntar al pop-up de Toon Boom, la pantalla muestra elementos digitales flotantes: extensiones 3D (hechas en Maya/Blender), efectos de partículas o variaciones de iluminación que complementan la maqueta de papel.

- **Modo B: Realidad Virtual (Inmersión Total)**

En la esquina inferior de la misma página web, A-Frame coloca automáticamente un botón con el icono de unas gafas de VR.<br> 
  - La Transición: Si el usuario pulsa ese botón (o coloca su celular dentro de un visor tipo Google Cardboard), la cámara del mundo real se apaga y el usuario es **"teletransportado"** adentro de la escena.

  - El Entorno: El alumno debe programar un entorno virtual de **360 grados (Skybox)** donde el usuario puede girar la cabeza y explorar la casa o el escenario tridimensional a escala real, como si hubiera encogido y entrado al libro.

# Directrices Técnicas para el Repositorio de GitHub

Para que los alumnos puedan programar esto sin complicarse con múltiples sitios web, enséñales a estructurar su proyecto en GitHub usando parámetros en la URL o rutas limpias. Por ejemplo, pueden tener un solo código base donde controlan qué página del libro se está leyendo:

- **Página 1 (La Casa):** ...github.io/libro-popup/?page=casa

- **Página 2 (El Carro):** ...github.io/libro-popup/?page=carro

En el código de JavaScript/A-Frame, **un simple condicional** cargará los **assets tridimensionales de Blender** o las secuencias de **Toon Boom** correspondientes a esa página específica.
