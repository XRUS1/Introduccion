# Entrega del proyecto final

¡Es una idea didáctica espectacular! Como proyecto de cátedra, esta propuesta es perfecta porque obliga a los estudiantes a resolver problemas en **tres dimensiones simultáneas:** el espacio físico (el papel), el espacio digital (la pantalla) y la variable temporal (la animación y la interacción).Para llevarlo a la práctica sin que los alumnos se frustren con la curva de aprendizaje técnico, aquí tienes una propuesta de rúbrica, alcance de proyecto y herramientas diseñada especialmente para el aula.

## 1. El Alcance Ideal para el Semestre

Para que el proyecto sea viable en un semestre, lo ideal es delimitar el alcance a un diorama de varias página o un folleto desplegable. El libro debe considerar dos integrantes para la Ingeniería de Papel; por los tiempos de secado, corte e ingeniería de papel.

## 2. Flujo de Trabajo Sugerido para los Alumnos en la Fase Análoga (Ingeniería de Papel)

Los alumnos deben prototipar primero el pop-up físico usando cartulinas blancas. Deben dominar los dos mecanismos básicos: el pliegue en **V (V-fold)** y las paralelas/vías **(Parallel folds)**.Fase Gráfica y Texturizado: **Diseñar la gráfica que irá impresa en el papel**. Esta gráfica impresa debe ser rica en contrastes y detalles visuales, ya que servirá como el **"marcador" (Target)** que la cámara del celular va a reconocer.<br>
- Fases posteriores:
    - **Fase Digital (Animación e Interacción):** Crear los elementos virtuales (modelos 3D, partículas de fuego, menús UI o audio) en la computadora. (2 integrantes)
    - **Fase de Integración:** Programar la lógica para que, al detectar el objeto físico, se despliegue el contenido digital interactivo. (2 integrantes)

## Herramientas utilizadas en la fase de Ingeniería de Papel

**Toon Boom Harmony** es el Ferrari de la animación 2D a nivel mundial. El hecho de que los alumnos dominen este software, sumado a Maya y Blender, abre una oportunidad pedagógica increíble para un proyecto de diseño interactivo.<br>

**Toon Boom Harmony** destaca por su potente motor de deformadores (**deformers**), **cinemática inversa** y **animación basada en nodos**. Esto permite generar proyectos con una estética híbrida impecable (mezclando arte tradicional 2D estilizado con la profundidad del 3D).Para integrar las animaciones hechas en **Toon Boom Harmony** dentro de la pila técnica de **WebXR, Three.js/A-Frame y GitHub**, los alumnos deben utilizar flujos de trabajo específicos para la web.


## Herraminetas utilizadas en la Fase Digital

**Maya, Blender** Significa que los alumnos tienen el perfil técnico ideal para desarrollar un proyecto de altísima calidad visual y espacial, ya que el **modelado 3D y el texturizado suelen ser los mayores cuellos de botella en proyectos de Realidad Aumentada (RA)**.

## Técnicas para el perfilado de proyecto

Al saber usar software **3D y 2D profesional**, debes poner especial atención en estos tres puntos técnicos durante las revisiones:

- **Optimización de Polígonos (Low Poly):** Los modelos creados en Maya o Blender se renderizarán en tiempo real en los celulares o tablets de los usuarios. Exígeles que trabajen con **mallas optimizadas (Low Poly)** y que utilicen técnicas de Baking (cocinado de texturas y luces) desde **Blender/Maya** hacia sus mapas 2D para que el rendimiento en el dispositivo móvil sea fluido.

- **Diseño del Marcador (Target) en 2D:** El software 2D no solo sirve para decorar el papel. Explícales que los algoritmos de RA buscan puntos de alto contraste y esquinas definidas para fijar el **objeto virtual**. Un diseño de papel con **gradientes suaves o formas circulares fallará en el rastreo;** necesitan diseños con patrones geométricos, texturas ricas o líneas contrastantes.

- **Formatos de Exportación:** Recuordar exportar sus animaciones desde Maya/Blender en formato .FBX (si usan Unity) o .GLB / .GLTF , asegurándose de **incluir las animaciones incrustadas** (Bake Animation)

## Herraminetas utilizadas en la fase de Integración

Abordar **WebXR, Three.js y A-Frame** en la web es la decisión más acertada para un entorno educativo moderno: elimina la barrera de que los alumnos tengan que publicar en App Stores, es **multiplataforma** por naturaleza y les enseña cómo funciona la arquitectura de la **web 3D** actual.<br>

Dado que el corazón de un libro pop-up interactivo depende del **Rastreo de Imágenes** (Image Tracking / Marker Tracking), el ecosistema WebXR nativo de los navegadores tiene un gran reto: la API oficial de **WebXR Image Tracking sigue considerándose experimental o en estado "Draft"** en los estándares de la W3C y [los navegadores móviles estándar.Por lo tanto, la forma óptima de resolver este proyecto en la web utilizando el código de código abierto disponible en **GitHub** es a través de capas de abstracción. Aquí tienes las dos rutas técnicas ideales que puedes plantear a tus estudiantes:

### Ruta 1: A-Frame + AR.js (El enfoque HTML declarativo y rápido)
Esta es la ruta ideal si se requiere que los alumnos se concentren rápidamente en el diseño de interacción, la composición del **pop-up** y la **importación de sus modelos 3D sin hundirse semanas enteras programando** la lógica de la cámara web. **A-Frame** es una biblioteca construida encima de **Three.js** que permite estructurar **mundos 3D** usando etiquetas estilo **HTML** (<a-scene>, ). Al complementarse con **AR.js**, obtiene un motor de rastreo de imágenes (**NFT - Natural Feature Tracking**) muy robusto que corre perfectamente en Safari de iOS y **Chrome** de Android sin configuraciones avanzadas.Estructura del código base en **GitHub** (Fácil de clonar y usar):Los alumnos pueden crear un archivo **index.html** básico directo en su repositorio de GitHub y servirlo mediante **GitHub Pages** (lo que les resuelve el requerimiento obligatorio de un servidor seguro HTTPS para activar la cámara web):
```html
<!DOCTYPE html>
<html>
  <head>
    <!-- Importar A-Frame y las librerías de AR.js -->
    <script src="https://aframe.io"></script>
    <script src="https://githack.com"></script>
  </head>
  <body style="margin: 0px; overflow: hidden;">
    
    <!-- Configuración de la escena WebAR con la cámara del celular activa -->
    <a-scene embedded arjs="sourceType: webcam; debugUIEnabled: false;">
      
      <!-- El marcador NFT: la imagen del pop-up físico que tus alumnos exportaron en software 2D -->
      <a-nft
        type="nft"
        url="ruta-al-servidor/descriptor-de-imagen"
        smooth="true"
        smoothCount="10"
        smoothTolerance=".01"
        smoothThreshold="5">
          
          <!-- Modelo 3D animado de Blender/Maya en formato .gltf exportado por el alumno -->
          
          

      </a-nft>
      
      
    </a-scene>
  </body>
</html>
```

**Nota para tus alumnos:** AR.js requiere procesar las imágenes 2D del pop-up para generar "descriptores" de rastreo. Para ello, deberán usar herramientas de código abierto en GitHub como el AR.js NFT Marker Generator, donde suben la imagen 2D de la página de su libro y descargan los archivos de rastreo necesarios.<br>
