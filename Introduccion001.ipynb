# Diseño Interactivo
**El diseño interactivo** es una disciplina que se enfoca en crear productos, servicios y experiencias digitales o físicos que permiten una **comunicación de doble vía entre el usuario y el sistema**. A diferencia del diseño estático tradicional, su objetivo principal es generar un diálogo significativo, donde las acciones de la persona (hacer clic, deslizar, hablar, moverse) modifican el comportamiento del contenido en tiempo.

## Pilares fundamentales.

Para lograr que esta comunicación fluya de forma natural, los diseñadores se basan en varios principios clave:
- **Usabilidad**: Garantiza que el producto sea fácil de usar, eficiente y amigable.
- **Consistencia**:Utiliza patrones visuales y de comportamiento repetitivos para que el usuario no deba reaprender las reglas a cada momento.
- **Previsión**: Organiza los elementos para que el usuario pueda anticipar qué pasará al realizar una acción.
- **Reducción de carga cognitiva**: Simplifica los procesos visuales e informativos para evitar saturar la mente del usuario.
  
## Áreas comunes de aplicación

Esta disciplina combina herramientas de arte, tecnología, comunicación y programación en sectores muy diversos:
- **Páginas web y apps móviles**: Creación de menús intuitivos, botones con respuestas animadas y formularios ágiles.
- **Videojuegos**: Estructuración de mecánicas de juego, interfaces y la respuesta del entorno virtual a los mandos del jugador.
- **Realidad Virtual (RV) y Aumentada (RA)**: Desarrollo de entornos inmersivos que reaccionan a los movimientos del cuerpo o superponen datos en el mundo real.
- **Instalaciones interactivas**: Espacios físicos (como museos o exposiciones) que cambian sus luces, sonidos o proyecciones cuando la gente camina o interactúa con ellos.

# Libro de realidad aumentada
Un libro de realidad aumentada (RA) es un ejemplo perfecto de **diseño interactivo de vanguardia**.<br>
Este formato rompe por completo la lectura pasiva tradicional. El diseñador debe planificar no solo cómo se ve la página impresa, sino cómo reacciona el contenido digital cuando el usuario interactúa con ella a través de una pantalla o un dispositivo óptico.

## ¿Por qué se considera diseño interactivo?

- **Comunicación bidireccional**: El libro físico actúa como un detonante físico. Cuando el lector apunta con su cámara, el sistema **responde mostrando modelos 3D**, **animaciones**, **sonidos o videos**
  
- **Control en manos del usuario**: El lector decide el ritmo. Puede girar el libro para ver el objeto virtual desde otros ángulos, acercarse para ver detalles o tocar la pantalla para activar funciones ocultas.

- **Fusión de entornos**: Conecta el diseño editorial clásico (tipografía, maquetación, papel) con el diseño de interfaces digitales (UI) y de experiencia de usuario (UX).

- **Retroalimentación inmediata**: Cada acción física del usuario produce una respuesta digital instantánea, que es la regla de oro del diseño interactivo.

# libros pop-up (o libros desplegables)

También se les conoce formalmente como **libros móviles o piezas de ingeniería de papel**, ya que utilizan **complejos mecanismos de pliegues y pestañas** para que las **figuras tridimensionales cobren vida al abrir cada página**.

# Combinación de Libros Pop-Up y Libros Realidad Aumentada. 

A esta fusión se le conoce como libros **pop-up de realidad aumentada** (o proyectos de realidad mixta aplicada a la literatura). <br>
Al unirlos, el impacto visual y la inmersión se multiplican exponencialmente porque se mezclan dos tipos de tridimensionalidad: la física (el papel que se levanta) y la digital (el contenido virtual que flota en la pantalla).

## ¿Cómo funciona esta combinación?

- **El papel sirve como "Ancla" (Target)**: En lugar de usar códigos QR o imágenes planas para activar la realidad aumentada, la aplicación móvil utiliza la propia forma física del pop-up (la casa o el carro de papel) como la señal de activación.
  
- **Modelado complementario**: Al abrir la página, se despliega un castillo de papel (pop-up). Cuando se mira a través de la tablet o celular, la realidad aumentada añade texturas dinámicas como nubes moviéndose alrededor de las torres, dragones volando que salen de las ventanas físicas o agua digital corriendo por el foso del castillo.


## Ventajas de combinarlos en el Diseño Interactivo

- **Interactividad háptica y digital**: El usuario experimenta el libro con el tacto (abriendo las solapas, sintiendo la textura del papel) al mismo tiempo que interactúa con la pantalla mediante toques o comandos de voz

- **Efectos temporales y narrativos**: Un libro pop-up físico es estático una vez abierto. La capa de realidad aumentada le permite al diseñador añadir el factor tiempo: mostrar el ciclo del día y la noche sobre la misma estructura de papel, o hacer que los personajes caminen de una habitación física a otra.

- **Potencial educativo masivo**

# Referencias de diseño Pup-UP y realidad aumentada

## 1. Acqua Alta (de Adrien M & Claire B)

Este es uno de los referentes internacionales más artísticos y reconocidos de esta fusión. Creado por los artistas franceses Adrien Mondot y Claire Bardainne, Acqua Alta es un **libro pop-up** que narra una historia sobre una inundación.
  https://www.dosdoce.com/2020/01/09/libros-entre-la-realidad-aumentada-y-los-pop-up/
  <br>
  
- **Cómo funciona:** Al abrir las páginas blancas y negras, se despliegan estructuras de papel minimalistas. Cuando miras el libro a través de una tablet o smartphone usando su app dedicada, figuras humanas virtuales interactúan directamente con los pliegues de papel, escalando los dobleces y nadando entre las formas físicas.

## 2. Simone Rocha x H&M Augmented Reality Pop-Up Book (por UNIT9)

Un ejemplo de alta costura y marketing interactivo que llevó el formato a los adultos. La agencia de innovación tecnológica **UNIT9**
  https://unit9.com/project/augmented-reality-pop-book/ <br>
creó un libro físico de edición limitada para el lanzamiento de una colección de ropa

## 3. "Augmented Stories" (de Helen Papagiannis)

Considerado históricamente como uno de los primeros experimentos del mundo en combinar estas dos disciplinas. <br> Desarrollado por la investigadora y artista canadiense Helen Papagiannis, el proyecto exploró el uso de software de reconocimiento de imágenes en dispositivos móviles aplicado a la ingeniería de papel tradicional. Puedes leer sobre sus inicios en su blog oficial de Augmented Stories.<br>

https://www.augmentedstories.com/blog/2019/2/5/first-ar-pop-up-book-for-ipad-and-iphone-using-image-recognition

## 4. Bookar (por ArtygenSpace)

En el sector educativo infantil, la empresa **ArtygenSpace** ha desarrollado el concepto de "Book Plus" o Bookar.

https://www.youtube.com/watch?v=MSdSBIcSdiY<br>

- **Cómo funciona:** Se trata de una plataforma que convierte libros ilustrados y pop-up en teatros tridimensionales interactivos. Utilizan "marcadores de papel" dinámicos donde los niños no solo ven la animación sobre el papel, sino que pueden arrastrar elementos en la pantalla para cambiar el curso de la historia o pintar sobre el libro y ver sus dibujos cobrar vida en 3D.
