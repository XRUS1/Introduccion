# El software de la Insta 360 X5

El software de la cámara genera esas **curvas automáticamente**, pero con un matiz crucial. Lo que el software genera matemáticamente en la imagen plana se llama **"proyección equirrectangular"**, la cual deforma las líneas rectas en curvas caprichosas para poder **"aplanar"** una esfera perfecta.

## 1. ¿Cómo genera el software esas curvas?

Cuando los alumnos toman una captura con la cámara, el **software nativo (Insta360 Studio)** procesa los dos lentes opuestos de 180° y **realiza un algoritmo de stitching (cosido)**. El resultado final que exporta es una **imagen plana (en proporción 2:1)** donde:

- El centro de la imagen se ve relativamente normal.

- Los polos (el techo/cielo y el suelo/piso) se estiran y se deforman en curvas caprichosas muy pronunciadas.<br>
  
Esta distorsión matemática es idéntica a cómo un mapamundi de Mercator deforma los continentes en los polos de la Tierra. El software de la cámara te entrega esa distorsión matemática lista, perfecta y sin errores de unión en un archivo .jpg

## 2. ¿Cómo se usan esas curvas en Blender y Maya?

Los alumnos, al ser artistas 3D, entenderán de inmediato el poder de esta imagen curva:

- Entornos **HDRI**: Si cargan ese .jpg curvo en el motor de render de Blender (Cycles) o Maya (Arnold) como una textura de entorno (Environment Texture), el software de modelado "desenvuelve" la imagen de forma inversa.

- Resultado: Las curvas caprichosas desaparecen mágicamente dentro del visor 3D y se convierten en un espacio tridimensional perfecto. Los reflejos en las ventanas, metales o plásticos de los objetos que tus alumnos modelen se adaptarán a la perfección a la iluminación del mundo real capturado.

## 3. El "Efecto Mágico" en WebXR (A-Frame y Three.js)

Cuando lleves este archivo al código web, pasarán exactamente por el mismo proceso matemático en tiempo real:Al declarar la etiqueta <a-sky src="#mi-foto-curva"></a-sky> en A-Frame, el framework le ordena a Three.js que cree una esfera tridimensional virtual (un **Skydome**), invierta sus caras para que apunten hacia adentro, y mapee textualmente esa imagen plana y curva en las coordenadas UV de la esfera.Cuando el usuario final abre el proyecto en GitHub Pages y se pone el visor de Realidad Virtual:

- La GPU del visor procesa esa textura curva.

- El usuario ve un espacio tridimensional perfectamente rectilíneo y realista, sin distorsiones.

-  Las "curvas caprichosas" se convierten en las esquinas de una habitación real, el horizonte de una montaña o el techo de un laboratorio.


# Horneado de Entornos (Environment Baking) o Renderizado Equirrectangular

Es una técnica de **nivel profesional** muy utilizada en la industria. Se le conoce formalmente como Horneado de Entornos (Environment Baking) o Renderizado Equirrectangular, y **Unity cuenta con herramientas nativas perfectas para esto**. <br>

El flujo de trabajo consiste en usar Unity como un "estudio de fotografía virtual" intermedio: **importas los recursos de la Insta360 X5**, colocas los modelos **3D complejos de Blender o Maya**, y luego **usas una cámara especial en Unity para grabar una nueva foto o video 360°** combinada, lista para WebXR.

## La Herramienta Clave en Unity: Recorder e Interactive Camera
 Para lograr esto, tú y tus alumnos necesitarán instalar el paquete oficial Unity Recorder (disponible de forma gratuita desde el Package Manager).<br>
 
### Paso 1: Configurar la Escena Híbrida en Unity

1. Creas una escena en Unity.
   
2. Configuras la foto/video de la Insta360 X5 como el Skybox de la escena (para que aporte el fondo y la luz del mundo real).

3. Importas los modelos 3D (personajes, efectos de partículas de Unity, luces digitales, interfaces de usuario). Los colocas en el espacio simulando que coexisten con la foto real.

### Paso 2: Configurar la Captura 360

1. En Unity, vas a la pestaña Window > General > Recorder > Recorder Window.

2. Añades un nuevo renderizador seleccionando Movie (para video 360) o Image Sequence (para fotos 360).

3. En la configuración de la cámara de captura (Source), cambias de Main Camera a 360 View.

4. El secreto técnico: Defines la resolución de salida (puedes elegir 4K u 8K) y el formato de salida como Equirectangular (el mismo formato de curvas caprichosas que entiende A-Frame).


### Paso 3: El Horneado (Baking)

Le das al botón **"Play"** en el Recorder. Unity procesará la escena y grabará un archivo .mp4 o .jpg en 360 grados que fusiona perfectamente el fondo real de la cámara con los elementos interactivos tridimensionales creados por tus alumnos.

## ¿Por qué esta técnica es una joya pedagógica para tu curso de WebXR?

Hacer este puente por **Unity** antes de ir a **WebXR** tiene tres ventajas:

1. **Optimización Extrema para Web:** WebXR a veces sufre para renderizar millones de polígonos o luces en tiempo real (especialmente en móviles o visores independientes). Con esta técnica, los alumnos pueden modelar un escenario hiperrealista y masivo en Maya, meterlo a Unity, y "hornearlo" en un solo video 360. Para la web, el navegador solo tendrá que reproducir un video plano (muy ligero), pero el usuario en VR sentirá que está dentro de un entorno ultra complejo.<br>

2. **Efectos Visuales Avanzados (VFX)**: Pueden usar los sistemas de partículas de Unity (fuego, humo, magia, lluvia) o físicas complejas, grabarlas en el video 360, y desplegarlas en A-Frame sin que consuma nada de procesador en el visor VR.

3. **Inclusión de Audio Espacial:** Unity permite exportar estos videos con metadatos de audio de 360 grados (Ambisonics). Cuando lo monten en la web, el sonido cambiará dependiendo de hacia dónde gire la cabeza el usuario.





# El Paso Final: Despliegue en A-Frame sobre GitHub

Una vez que Unity te entrega ese nuevo video **horneado en 360 con todo integrado**, el código en tu stack de **WebXR** cambia ligeramente para reproducir video en lugar de una foto estática:

```js
<a-scene>
  <a-assets>
    <!-- El video 360 horneado en Unity y subido a GitHub -->
    <video id="video-unity-360" autoplay loop crossOrigin="anonymous" src="./assets/escena_horneada_unity.mp4"></video>
  </a-assets>

  <!-- En lugar de a-sky con imagen, usamos el video como fondo envolvente -->
  <a-videosphere src="#video-unity-360" rotation="0 -90 0"></a-videosphere>

  
</a-scene>
```

Esta técnica cierra el círculo perfecto de tu stack: **Insta360 (Captura) ➔ Blender/Maya (Assets) ➔ Unity (Composición y Horneado 360) ➔ GitHub/A-Frame (Despliegue WebXR)**.


# Horneado por geometría física (SkyDom)

Para lograr ese flujo, la técnica consiste en **invertir las caras de una esfera** (hacerle un Flip Normals) a través de un código de sombreador (Shader). De forma nativa, **las esferas de Unity se construyen para ser vistas desde afuera;** al **invertir las caras con un Shader**, la esfera **se vuelve visible únicamente desde adentro**, transformándose en un Skydome físico.<br>

Luego, puedes colocar una cámara en el centro de esa esfera en Unity, capturar el render equirrectangular y exportarlo directamente hacia tu stack de WebXR / A-Frame.<br>

## Código del Shader para Skydome (Unity)

Para crear este sombreador en Unity, haz clic derecho en tu ventana de proyecto: Create > Shader > Unlit Shader, nómbralo SkydomeInvertido y pega el siguiente código:<br>

```hlsl
Shader "Custom/SkydomeInvertido"
{
    Properties
    {
        // El inspector de Unity nos pedirá la foto 360 de la Insta360
        _MainTex ("Textura de Entorno 360 (LatLong)", 2D) = "white" {}
    }
    SubShader
    {
        Tags { "RenderType"="Opaque" "Queue"="Geometry" }
        LOD 100

        // EL SECRETO TÉCNICO: Cull Front le dice a Unity que dibuje 
        // únicamente las caras INTERNAS de la esfera y oculte las externas.
        Cull Front
        ZWrite On

        Pass
        {
            CGPROGRAM
            #pragma vertex vert
            #pragma fragment frag
            #include "UnityCG.cginc"

            struct appdata
            {
                float4 vertex : POSITION;
                float2 uv : TEXCOORD0;
            };

            struct v2f
            {
                float2 uv : TEXCOORD0;
                float4 vertex : SV_POSITION;
            };

            sampler2D _MainTex;
            float4 _MainTex_ST;

            v2f vert (appdata v)
            {
                v2f o;
                o.vertex = UnityObjectToClipPos(v.vertex);
                
                // INVERSIÓN DE LA TEXTURA: 
                // Al mirar desde adentro, la imagen se vería espejada (al revés). 
                // Al restar la coordenada X a 1, corregimos el efecto espejo.
                o.uv = v.uv;
                o.uv.x = 1.0 - o.uv.x; 
                
                return o;
            }

            fixed4 frag (v2f i) : SV_Target
            {
                // Mapea y renderiza el color de la textura directamente sin sombras
                fixed4 col = tex2D(_MainTex, i.uv);
                return col;
            }
            ENDCG
        }
    }
}
```

## Cómo implementarlo en Unity (Paso a Paso)

Una vez que tengas el código listo, el flujo de trabajo en la clase para armar el Skydome intermedio es este:

1. **Crear el Material:** Haz clic derecho sobre el archivo del Shader en Unity y selecciona Create > Material. Automáticamente se creará un material que usa este sombreador inverso.

2. **Asignar la textura:** Arrastra la foto de la Insta360 X5 al recuadro de textura (_MainTex) en el inspector de ese material.

3. **Crear la geometría:** En la jerarquía de la escena, crea una esfera física: GameObject > 3D Object > Sphere.

4. **Escalar la esfera:** Ponle una escala gigante (por ejemplo: X: 50, Y: 50, Z: 50) para que envuelva por completo todo lo que tus alumnos vayan a modelar. Asegúrate de que la posición de la esfera esté en el origen absoluto (0, 0, 0).

5. **Aplicar el material:** Arrastra tu nuevo material a la esfera. Verás que desde afuera de la esfera el objeto parece "invisible" o transparente, pero si metes la cámara de Unity al centro (0, 0, 0), estarás dentro de un domo fotográfico perfecto de 360 grados.

## Pasar de esta Esfera de Unity a WebXR

Una vez que tus alumnos tienen montado este Skydome en Unity con sus assets de Blender/Maya acomodados dentro, la transición hacia el ecosistema web es directa:

- **Para una escena fija:** Utilizan el componente **Unity Recorder (en su modo 360 View)** para capturar una única foto de lo que ocurre dentro del domo e importarla directamente en la etiqueta <a-sky> de A-Frame.

- **Para un fondo interactivo o animado**: Pueden animar luces digitales, efectos de partículas o personajes dentro del Skydome de Unity, exportar un video .mp4 en 360 y reproducirlo en WebXR usando la etiqueta <a-videosphere>, logrando entornos dinámicos sumamente inmersivos sin penalizar el rendimiento del navegador.

## SkyDom es un proceso de composición y horneado (baking) 360, pero realizado a través de geometría física en lugar de configuraciones de cámara globales.

Al meter la foto de la Insta360 X5 dentro del shader de la esfera y colocar los modelos 3D al centro, estás haciendo **composición espacial**. Cuando usas la cámara de captura para exportar el resultado final, estás horneando toda esa escena en un **nuevo archivo plano equirrectangular** listo para tu stack de WebXR


| Criterio | Método A: Skybox Global (Render Settings) | Método B: Skydome Físico (Esfera + Shader) |
| :--- | :--- | :--- |
| **Naturaleza** | El fondo es infinito.<br>No tiene posición real en el espacio 3D. | El fondo es un objeto 3D real.<br>Tiene escala y posición fija (ej. `0,0,0`). |
| **Iluminación** | Unity usa la foto para iluminar automáticamente los modelos 3D del alumno. | Los modelos se iluminan solo con las luces digitales colocadas dentro de la esfera. |
| **Control de Escala** | El usuario no puede moverse respecto al fondo.<br>Siempre está a la misma distancia. | Permite trucos visuales al alterar la escala de la esfera (hacerla ovalada o gigante). |
| **Facilidad de Composición** | Flujo más rápido para flujos de trabajo tradicionales de videojuegos. | Más intuitivo para artistas de Maya/Blender acostumbrados a mapear domos (*HDRI*). |
| **Flujo WebXR** | Se exporta directo a través de la cámara de captura 360 hacia GitHub. | Requiere invertir las normales de la esfera o usar un shader *Double Sided* antes de capturar. |


## La Gran Ventaja Pedagógica del Skydome Físico (Método B)

Para las personas que ya conocen el concepto de coordenadas UV y geometría, usar el método de la esfera con el shader modificado les otorga un control artístico superior por tres razones:

1. **Efectos de Postprocesado Locales:** Al ser una esfera física, se pueden aplicar efectos visuales (como niebla volumétrica o Shaders de distorsión de agua) que afecten selectivamente al espacio interior del domo, algo imposible con un Skybox global infinito.

2. **Máscaras y Recortes:** Pueden recortar pedazos de la esfera mediante código o texturas de transparencia (Alpha Clipping) para fusionar el mundo real capturado por la Insta360 X5 con un cielo procedural o un universo fantástico modelado por ellos.

3. **Simulación de Distancia (Paralaje)**: Si colocan elementos intermedios (como edificios lejanos en 3D de Blender) entre la cámara central y las paredes de la esfera, el horneado final 360 tendrá una profundidad visual mucho más realista y creíble al exportarse a WebXR.
